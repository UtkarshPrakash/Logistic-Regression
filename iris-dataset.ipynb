```python
iris = pd.read_csv("../input/iris-flower-dataset/IRIS.csv")
iris.info() # We can see there are no empty values
```

```python
iris.hist(linewidth = 1, edgecolor = 'black')
fig = plt.gcf()
fig.set_size_inches(10,5)

plt.show()
```

```python
X = iris[['sepal_length', 'sepal_width', 'petal_length', 'petal_width']]
y = iris['species']
print(X)
```

```python
train_x, test_x, train_y, test_y = train_test_split(X, y, train_size = 0.85)
```

```python
model = LogisticRegression()
model.fit(train_x, train_y)
```

```python
prediction = model.predict(test_x)
accuracy = metrics.accuracy_score(prediction, test_y)
print('Accuracy : ', accuracy) # We are getting pretty good accuracy
```
